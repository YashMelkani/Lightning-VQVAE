# Evaluate VQVAE

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import torch

from tqdm import tqdm

from utils.data import create_vqvae_dataloader
from utils.vqvae import ConditionalVQVAE, UnconditionalVQVAE
```

## Unconditional

```python
base_dir = os.path.join(os.environ["SCRATCH"], "Dr. Ju", "track-sort-vqvae", "[8, 13, 17]_6")
testset_config = {'batch_size': 256,
                  'events_range': [1110, 1120],
                  'track_len': 32,
                  'shuffle_files': False,
                  'shuffle_tracks': False }

test_loader = create_vqvae_dataloader(testset_config, base_dir=base_dir, conditional=False)
```

```python
device = 'cuda'

ckpt_path = "./results/unconditional-vqvae/lightning_logs/version_0/checkpoints/epoch=00094-val_loss=0.015.ckpt"
model = UnconditionalVQVAE.load_from_checkpoint(ckpt_path)

model.eval()
model.to(device=device)

print("Model loaded to", device)
```

```python
hits = []
recon_hits = []

for batch in tqdm(test_loader):
        
    hits.append(batch)
    
    vq_loss, recon, emb_idxs = model(batch.to(device=device))
    
    recon = recon.detach().cpu().numpy()
    
    recon_hits.append(recon)
    
hits = np.concatenate(hits, axis=0) * 200 # remove normalization
recon_hits = np.concatenate(recon_hits, axis=0) * 200 # remove normalization
```

```python
print('Total hits:', hits.shape[0])
```

```python
# create x, y, z, r reconstructed error plots
fig, ax = plt.subplots(1,3, figsize=(6,2), dpi = 200, sharey="row")

params = ["x", "y", "z"]
for i in range(3):
    d = hits[:, i] - recon_hits[:, i]
    ax[i].hist(d, bins = 50, range=(-100, 100))
    
    ax[i].set_xlabel(f"${params[i]} - {params[i]}_{{recon}}$ (mm)")
    
    ax[i].set_xlim(-100, 100)
    
ax[0].set_ylabel("Counts")
    
plt.tight_layout()
plt.show()
```

```python
error = np.linalg.norm(hits - recon_hits, axis = 1)

fig, ax = plt.subplots(1,2, figsize=(6,3), dpi = 200, sharex='row')

cnts, bins, _ = ax[0].hist(error, bins = 50, range=(0, 150), histtype='bar', alpha=0.5)
ax[0].set_ylabel("Counts")
ax[0].set_xlabel("Reconstruction Error (mm)")
ax[0].set_xlim(0, 150)
ax[0].set_title('Error Distribution')

cum, bins, _ = ax[1].hist(error, bins=bins, histtype='step', cumulative=True, density=True, alpha=0.8)
ax[1].set_xlabel("Reconstruction Error (mm)")
ax[1].set_ylim(0, 1)
ax[1].set_title('Cumulative Distribution')

bw = bins[1] - bins[0]
auc = cum.sum() * bw / 150
print('CDF AUC:', auc)

plt.tight_layout()
plt.show()
```

```python
# percent of hits reconstructed w/ < 25mm error
np.count_nonzero(error < 25)  / len(error)
```

## Conditional

```python
base_dir = os.path.join(os.environ["SCRATCH"], "Dr. Ju", "track-sort-vqvae", "[8, 13, 17]_6")
testset_config = {'batch_size': 256,
                  'events_range': [1110, 1120],
                  'track_len': 32}

test_loader = create_vqvae_dataloader(testset_config, base_dir=base_dir, conditional=True)
```

```python
device = 'cuda'

ckpt_path = "./results/conditional-vqvae/lightning_logs/version_0/checkpoints/epoch=00099-val_loss=0.001.ckpt"
model = ConditionalVQVAE.load_from_checkpoint(ckpt_path)

model.eval()
model.to(device=device)

print("Model loaded to", device)
```

```python
def move_2_device(tensors, device='cpu'):
    
    out = []
    for tensor in tensors:
        if tensor is None:
            out.append(None)
        else:
            out.append(tensor.to(device=device))
    
    return out

n_tracks = 0
hits = []
recon_hits = []

for batch in tqdm(test_loader):
    
    tracks, pad_masks = batch
    
    n_tracks += tracks.shape[0]
    hits.append(tracks[~pad_masks])
    
    track_length = tracks.shape[1]
    att_mask = torch.triu(torch.ones(track_length, track_length, dtype=torch.float32) * float('-inf'), diagonal=1)
    
    tracks, pad_masks, att_mask = move_2_device([tracks, pad_masks, att_mask], device=device)
    
    vq_loss, recon, emb_idxs = model(tracks, pad_masks, att_mask=att_mask)
    
    recon = recon.detach().cpu().numpy()
    pad_masks = pad_masks.detach().cpu().numpy()
    
    recon_hits.append(recon[~pad_masks])
    
hits = np.concatenate(hits, axis=0) * 200
recon_hits = np.concatenate(recon_hits, axis=0) * 200
```

```python
print('Total hits:', hits.shape[0])
print('Total tracks:', n_tracks)
```

```python
# create x, y, z, r reconstructed error plots
fig, ax = plt.subplots(1,3, figsize=(6,2), dpi = 200, sharey="row")

params = ["x", "y", "z"]
for i in range(3):
    d = hits[:, i] - recon_hits[:, i]
    ax[i].hist(d, bins = 50, range=(-100, 100))
    
    ax[i].set_xlabel(f"${params[i]} - {params[i]}_{{recon}}$ (mm)")
        
    ax[i].set_xlim(-100, 100)
    
ax[0].set_ylabel("Counts")
    
plt.tight_layout()
plt.show()
```

```python
error = np.linalg.norm(hits - recon_hits, axis = 1)

fig, ax = plt.subplots(1,2, figsize=(6,3), dpi = 200, sharex='row')

cnts, bins, _ = ax[0].hist(error, bins = 50, range=(0, 150), histtype='bar', alpha=0.5)
ax[0].set_ylabel("Counts")
ax[0].set_xlabel("Reconstruction Error (mm)")
ax[0].set_xlim(0, 150)
ax[0].set_title('Error Distribution')

cum, bins, _ = ax[1].hist(error, bins=bins, histtype='step', cumulative=True, density=True, alpha=0.8)
ax[1].set_xlabel("Reconstruction Error (mm)")
ax[1].set_ylim(0, 1)
ax[1].set_title('Cumulative Distribution')

bw = bins[1] - bins[0]
auc = cum.sum() * bw / 150
print('CDF AUC:', auc)

plt.tight_layout()
plt.show()
```

```python
# percent of hits reconstructed w/ < 25mm error
np.count_nonzero(error < 25)  / len(error)
```
